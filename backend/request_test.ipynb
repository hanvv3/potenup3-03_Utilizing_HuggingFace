# 1. get endpoint 테스트

```python
import requests

fastapi_url = 'http://localhost:8080'

response = requests.get(fastapi_url)
print(response.status_code)
```

```python
print(response.json())
```

# 2. upload_image 테스트

```python
upload_image_url = f'{fastapi_url}/detect_image'
print(upload_image_url)
```

```python
with open('../images/test_image.jpg', 'rb') as f:
    files = {
        'file': f
    }
    response = requests.post(
        upload_image_url,
        files=files,
    )
    if response.status_code == 200:
        print(response.json())
```
