# 1. 모델 불러오기

```python
# 라이브러리 설치: uv add ultralytics
from ultralytics import YOLO

model = YOLO("models/yolo26n.pt")    # 다운로드 위치
model
```

# 2. 이미지 예측하기

```python
results = model(
    source='./images/yolo_image.jpg',
    save=True
)
results
```

```python
results[0].boxes.xywh
```

```python
print(f"cls  : {results[0].boxes.cls[3]}")
print(f"conf : {results[0].boxes.conf[3]}")
print(f"data : {results[0].boxes.data[3]}")

print(f"cls name : {results[0].names[results[0].boxes.cls[3].item()]}")
```
