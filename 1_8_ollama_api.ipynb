# 2. Generate a Response

```python
import requests 

url = "http://localhost:11434/api/generate"
data = {
    "model": "gemma3:4b",
    "prompt": "네 소개를 200자 이내로 해줘."
}

response = requests.post(
    url=url,
    json=data
)
if response.status_code == 200:
    print(response.text)
```

```python
import json 

with requests.post(url, json=data) as r:
    for line in r.iter_lines():
        # print(line)
        # print(type(line))
        response_data = json.loads(line)
        # print(response_data)
        # print(type(response_data))
        text = response_data["response"]
        print(f"[응답받은 텍스트]: {text}")
        if text:
            print(text, end="")
```

# 3. Generate a chat message

```python
import requests

url = 'http://localhost:11434/api/chat'
data = {
    'model': 'gemma3:4b',
    'messages':[
        {'role': 'system', 'content': '당신은 불친절한 사람입니다.'},
        {'role': 'user', 'content': '안녕?'}
    ],
    'stream': False
}

response = requests.post(
    url=url,
    json=data
)

if response.status_code == 200:
    response_data = response.json()
    print(response_data)
```

```python
response_data['message']
```

```python
import requests 

url = "http://localhost:11434/api/chat"
data = {
    'model': 'deepseek-r1:8b',
    'messages':[
        {'role': 'system', 'content': '당신은 불친절한 사람입니다.'},
        {'role': 'user', 'content': '안녕?'}
    ],
    'stream': False
}

response = requests.post(
    url=url,
    json=data
)
if response.status_code == 200:
    response_data = response.json()
    print('[Thinking]')
    print(response_data['message']['thinking'])
    print('[Response]')
    print(response_data['message']['content'])
```
